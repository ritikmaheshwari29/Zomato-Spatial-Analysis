```python
pip install geopy
```

```python
from geopy.geocoders import Nominatim 
```

```python
geolocator=Nominatim(user_agent='app')
```

```python
lat=[]
long=[]
for location in locations['Name']:
    location=geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        long.append(np.nan)
    else:
        lat.append(location.latitude)
        long.append(location.longitude)
```

```python
locations['lat'] = lat
locations['long'] = long
```

```python
def Heatmap_Zone(zone):
    df3=data[data['cuisines']==zone]
    df_zone=df3.groupby(['location'],as_index=False)['url'].agg('count')
    df_zone.columns=['Name','count']
    df_zone=df_zone.merge(locations,on="Name",how='left').dropna()
    HeatMap(df_zone[['lat','long','count']],zoom=20,radius=15).add_to(basemap)
    return basemap
```
